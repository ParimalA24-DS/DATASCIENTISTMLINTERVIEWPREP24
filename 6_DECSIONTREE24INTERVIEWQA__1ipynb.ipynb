**Depedent and indepednt features**

```python
X=df.iloc[:,:-1]#takefrom 2nd dataset
y=iris.target  #takefrom 1st dataset :iris
```

## **Traintest split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
y_train
```

## **I.DECISION TREECLASSIFIER:**
- Post pruning
- Pre pruning

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treem=DecisionTreeClassifier()
```

```python
treem.fit(X_train,y_train) #modelget train
```

```python
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(treem,filled=True)
plt.show()
```

```python
#prediction
y_pred=treem.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

**Post pruning**
- post pruning :cutting down the specific nodes after building the Decision Tree

```python
#post pruning :cutting down the specific nodes after building the Decision Tree
```

```python
## Postpruning
treemodel=DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(treemodel,filled=True)
```

```python
#prediction
y_pred1=treemodel.predict(X_test)
```

```python
score=accuracy_score(y_pred1,y_test)
print(score)
```

```python
print(classification_report(y_pred1,y_test))
```

**II.Pre pruning:**
- It is used before the construction of decision tree wih hyperparameter tuning and crsoss validation

- - criterion: Determines how the quality of a split is measured.
- - splitter: Determines the strategy used to choose the split at each node.
- - max_depth: Limits the depth of the tree to prevent it from becoming too complex and overfitting the data.
- - max_features: Limits the number of features considered for splitting at each node, which can reduce overfitting and improve computational efficiency.

```python
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
#GridSearchCV is used for hyperparameter tuning with cross validation
```

```python
treemodel2=DecisionTreeClassifier()
cv=GridSearchCV(treemodel2,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred2=cv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
score=accuracy_score(y_pred2,y_test)
score
```

```python
print(classification_report(y_pred2,y_test))
```
