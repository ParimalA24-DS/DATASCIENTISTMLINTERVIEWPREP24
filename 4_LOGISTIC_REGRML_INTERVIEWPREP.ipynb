### **INDEPENDENT AND DEPEDENT FEATURES:**

### **TRAIN_TEST_SPLIT**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# USE HYPERPARAMETER TUNIG USING GridSeachCV:
from sklearn.linear_model import LogisticRegression

cl=LogisticRegression()
```

```python
 from sklearn.model_selection import GridSearchCV #hyperparameter tunig
 parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,7,9,10,20,40,50],'max_iter':[100,20,300]}
parameter
# Hyperparameter tuning is the process of finding the best set of hyperparameters for a machine learning model
#to improve its performance.
```

- *Here*:  
-- The C hyperparameter is the inverse of regularization strength; smaller values specify stronger regularization.


```python
clcv=GridSearchCV(cl,param_grid=parameter,cv=5,scoring='accuracy') #hyperparameter tunig
```

```python
clcv.fit(X_train,y_train)
```

```python
#findbest parameter isselected
print(clcv.best_params_)
```

```python
#accuracyscore:
print(clcv.best_score_)
```

```python
#prediction test dataset :
y_predict=clcv.predict(X_test)
```

```python
#accuracy score?
from sklearn.metrics import accuracy_score,pair_confusion_matrix,classification_report
```

```python
 score=accuracy_score(y_predict,y_test)
 print(score)
```

```python
print(classification_report(y_predict,y_test))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
cm = confusion_matrix(y_test, y_predict)
cm
```
